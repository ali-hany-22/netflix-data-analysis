## **Cleaning**
### **DataType**

```python
netflix.head()
```

#### **date_added to datetime**

```python
netflix['date_added'] = pd.to_datetime(netflix['date_added'], errors='coerce')
netflix.info()
```

### **type          , rating        to category **

```python
print(f'value in type : {netflix["type"].unique()}')
print(f'value in rating : {netflix["rating"].unique()}')
```

```python
print(f'value in country : {netflix["country"].unique()}')
```

```python
category_columns = ['type', 'rating','country']
for col in category_columns:
    netflix[col] = netflix[col].astype('category')
```

```python
netflix.info()
```

### **duration**

```python
netflix['duration_num'] = netflix['duration'].str.extract('(\d+)').astype(float)
netflix['duration_type'] = netflix['duration'].str.extract('([a-zA-Z]+)').astype('category')
```

```python
netflix.info()
```

## **Missing Values**

### **cast,director --> Unknown**

```python
netflix['cast'].fillna('Unknown', inplace=True)
netflix['director'].fillna('Unknown', inplace=True)
```

### **country --> max**

```python
most_common_country = netflix['country'].mode()[0]
netflix['country'].fillna(most_common_country, inplace=True)
```

## **date_added --> dell**

```python
netflix = netflix[netflix['date_added'].notnull()]
```

## **rating --> max**

```python
most_common_rating = netflix['rating'].mode()[0]
netflix['rating'].fillna(most_common_rating, inplace=True)
```

## **duration_num , duration_type --> dell**

```python
netflix.dropna(subset=['duration_num', 'duration_type'], inplace=True)
```

```python
df = pd.read_csv('netflix_clean.csv')  # غيّر اسم الملف إذا لزم

# العنوان
st.title("📺 Netflix Dashboard")

# فلاتر
type_filter = st.multiselect("اختر نوع العرض", options=df['type'].unique(), default=df['type'].unique())
country_filter = st.multiselect("اختر الدولة", options=df['country'].dropna().unique(), default=df['country'].dropna().unique())

# تصفية البيانات حسب الفلاتر
df_filtered = df[(df['type'].isin(type_filter)) & (df['country'].isin(country_filter))]

# إحصائيات عامة
st.subheader("إحصائيات عامة")
col1, col2, col3 = st.columns(3)
col1.metric("عدد العروض", len(df_filtered))
col2.metric("عدد الدول", df_filtered['country'].nunique())
col3.metric("عدد التصنيفات", df_filtered['rating'].nunique())

# رسم بياني: عدد العروض حسب النوع
fig1 = px.histogram(df_filtered, x="type", color="type", title="عدد العروض حسب النوع")
st.plotly_chart(fig1)

# رسم بياني: التصنيفات الأكثر شيوعًا
rating_counts = df_filtered['rating'].value_counts().reset_index()
rating_counts.columns = ['rating', 'count']

fig2 = px.bar(rating_counts,
              x='rating', y='count',
              title="التصنيفات الأكثر شيوعًا",
              labels={'rating': 'التصنيف', 'count': 'عدد العروض'})
st.plotly_chart(fig2)

# رسم بياني: العروض حسب سنة الإصدار
fig3 = px.histogram(df_filtered, x="release_year", nbins=30, title="العروض حسب سنة الإصدار")
st.plotly_chart(fig3)

# أشهر الدول إنتاجًا
top_countries = df_filtered['country'].value_counts().head(10).reset_index()
top_countries.columns = ['country', 'count']

fig4 = px.pie(top_countries, names='country', values='count', title="أشهر الدول إنتاجًا للمحتوى")

# ملاحظة ختامية
st.markdown("---")
st.info("لوحة تفاعلية لبيانات Netflix. يمكنك تصفية العروض حسب النوع والدولة.")
```
